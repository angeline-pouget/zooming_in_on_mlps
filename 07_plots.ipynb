```python
from matplotlib import pyplot as plt
import numpy as np
```

### Adversarial Accuracies Plots

```python
acc_fgsm_mlp = np.load('adv_robustness/accuracy_fgsm_mlp_cifar10.npy')
acc_fgsm_cnn = np.load('adv_robustness/accuracy_fgsm_cnn_cifar10.npy')
acc_fgsm_vit = np.load('adv_robustness/accuracy_fgsm_vit_cifar10.npy')

top5_fgsm_mlp = np.load('adv_robustness/top5_fgsm_mlp_cifar10.npy')
top5_fgsm_cnn = np.load('adv_robustness/top5_fgsm_cnn_cifar10.npy')
top5_fgsm_vit = np.load('adv_robustness/top5_fgsm_vit_cifar10.npy')

eps = np.arange(0,0.26,0.0125)

plt.plot(eps, acc_fgsm_mlp, label='MLP Top-1', color='blue')
plt.plot(eps, top5_fgsm_mlp, label='MLP Top-5', color='lightblue')
plt.plot(eps, acc_fgsm_cnn, label='CNN Top-1', color='green')
plt.plot(eps, top5_fgsm_cnn, label='CNN Top-5', color='lightgreen')
plt.plot(eps, acc_fgsm_vit, label='ViT Top-1', color='red')
plt.plot(eps, top5_fgsm_vit, label='ViT Top-5', color='pink')

plt.legend()
plt.xlabel('Perturbation')
plt.ylabel('Adversarial Accuracy (%)')
plt.title('Adversarial Accuracy on CIFAR-10 for FGSM Attack')
plt.ylim(0,100)

plt.show()
```
